```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image

# Number of images in the grid
grid_size = 4
num_images = grid_size * grid_size

# Directory containing validation images
val_dir = './val'

# Get a list of image paths in the validation directory
image_paths = [os.path.join(val_dir, fname) for fname in os.listdir(val_dir) if fname.endswith(('.jpg', '.png'))]

# Select random images
selected_images = random.sample(image_paths, num_images)

# Initialize the figure
fig, axes = plt.subplots(grid_size, grid_size, figsize=(12, 12))
fig.suptitle("Random Validation Images with Labels", fontsize=16)

# Display images with their labels
for ax, img_path in zip(axes.flatten(), selected_images):
    # Open the image
    image = Image.open(img_path)
    
    # Extract the label from the filename (assuming filename format is label.<unique_id>.jpg/png)
    label = os.path.basename(img_path).split('.')[0]
    
    # Show the image and label
    ax.imshow(image)
    ax.set_title(label)
    ax.axis('off')

plt.tight_layout()
plt.subplots_adjust(top=0.92)  # Adjust top for the suptitle
plt.show()
```
